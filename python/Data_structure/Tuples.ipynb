### Introduction of Tuples
##### Tuples are ordered collections of items that are immutable. They are similar to lists,but their immutability makes them different.

```python
## Creating a tuple 
empty_tuple = ()
print(empty_tuple)
print(type(empty_tuple))
```

```python
lst = list()
print(type(lst))
tpl = tuple()
print(type(tpl))
```

```python
lst = list((1,2,3,4))
print(lst)
```
