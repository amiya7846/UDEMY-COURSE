### Introduction of Tuples
##### Tuples are ordered collections of items that are immutable. They are similar to lists,but their immutability makes them different.

```python
## Creating a tuple 
empty_tuple = ()
print(empty_tuple)
print(type(empty_tuple))
```

```python
lst = list()
print(type(lst))
tpl = tuple()
print(type(tpl))
```

```python
lst = list((1,2,3,4))
print(lst)
```

```python
numbers = tuple([1,2,3,4,5,6,7])
numbers
```

```python
list((1,2,3,4,5,6,7,8))
```

```python
mixed_tuple = (1,"Hello world",3.14,True)
print(mixed_tuple)
```

```python
##Accessing Tuple Elements 
numbers
```

```python
numbers[2]
```

```python
print(numbers[-1])
```

```python
numbers[::-1]
```

```python
## Tuple operations 
concatination_tuple = numbers + mixed_tuple
print(concatination_tuple)
```

```python
mixed_tuple * 3
```

```python
numbers * 3
```

```python
## Immutable Nature of Tuple 
## Tuples are immutable, meaning their elements cannot be changed once assigned.
lst = [1,2,3,4,5,6,7]
print(lst)
lst[1] = "Amiya"
print(lst)
```

```python
numbers[1] = "Amiya"
print(numbers) ##Tuple object doesnot support item assignment 
```

```python
##Tuple methods 
print(numbers.count(1)) ## This count method is used to count how many times the element is present in the tuple.
print(numbers.count(10))
```

```python
print(numbers.index(3)) ## This index method is used to find the index of the first occurrence of the specified element in the tuple.
##It will show that the element 3 is present in which index. If the element is not found, it will raise a ValueError.
print(numbers.index(10)) ## This will raise a ValueError because the element 10 is not present in the tuple.
```
