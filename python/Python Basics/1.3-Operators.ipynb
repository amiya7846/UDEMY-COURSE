```python
##Arithmetic Operators
a = 10
b = 15

add_result = a+b #Addition
sub_result = a-b #Subtraction
mul_result = a*b #Multiplication
div_result = a/b #Division
mod_result = a%b #Modulus
floor_div_result = a//b #Floor Division
exp_result = a**b #Exponentiation

print("Addition:", add_result)
print("Subtraction:", sub_result)
print("Multiplication:", mul_result)
print("Division:", div_result)
print("Modulus:", mod_result)
print("Floor Division:", floor_div_result)
print("Exponentiation:", exp_result)

```

### Logical Operators

- AND:-for and operators if the condition is true then only it will return true otherwise it will return false
- OR:-for or operators if any one condition is true then it will return true otherwise it will return false
- not operator will return the opposite of the condition if the condition is true then it will return false and if the condition is false then it will return true
