```python
## if statements 
age  = 20
if age >= 18:
    print("You are eligible to vote.")
    
```

```python
##else statements
## The else staement is used when the condition in the if staement is false.
age = 17
if age >= 18:
    print("You are eligible to vote.")
else:
    print("You are not eligible to vote.You are a minor.")
    
```

```python
##elif statements
## The elif statement is used to check multiple conditions.It stands for "else if".It is used when you have more than two options to choose from.
age = 17
if age < 13:
    print("You are a child.")
elif age < 18:
    print("You are a teenager.")
else:
    print("You are an adult.")        
```

```python
## Nested conditional statements
## Nested conditional statements are if statements that are inside another if statement. They are used to check multiple conditions in a more complex way.
## Number even, odd, negative, positive
num = int(input("Enter the Number:"))
if num >= 0:
    print("The number is positive")
    if num % 2 == 0:
        print("The number is even")
    else:
        print("The number is odd")
else:
    print("The number is negative")        
```

```python
## Practical Example
## Determine if a year is a leap year using nested condition statements.
year = int(input("Enter the year:"))
if (year % 4 ==0 and year % 100 != 0):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")    
```

LOOPS IN PYTHON 

```python
##For loops:- it is used to iterate over a sequence (list, tuple, string) or other iterable objects. It is used when you want to execute a block of code a certain number of times.
for i in range(5):
    print(i)
```

```python
for i in range(1,10,2):
    print(i)
```

```python
for i in range (10,1,-2):
    print(i)
```

```python
## Strings 
str =  "Amiya Swain"
for i in str:
    print(i)
```

```python
##While loops:= The while loops continues to execute a block of code as long as a specified condition is true. It is used when you want to repeat a block of code an unknown number of times until a certain condition is met.
count = 0
while count < 5:
    print(count)
    count += 1
```

```python
count = 0
while count%2==0:
    print(count)
    count += 1
```

```python
## Loop Control statements 
##break statement:- The break statement is used to exit a loop prematurely when a certain condition is met. It is used to stop the execution of a loop when a specific condition is satisfied.
for i in range (10):
    if i == 5:
        break
    print(i)
```

```python
## continue statement:- The continue statement is used to skip the current iteration of a loop and move on to the next iteration. It is used when you want to skip a certain condition in a loop.
for i in range (10):
    if i%2==0:
        continue
    print(i)
```

```python
## pass statements:- The pass statement is a null statement that is used as a placeholder in situations where a statement is syntactically required but no action is needed. It is often used in loops, functions, or classes that are not yet implemented.
for i in range (5):
    if i == 3:
        pass
    print(i)
```

```python
## Nested loops :- Nested loops are loops that are inside another loop. They are used to iterate over multiple sequences or to perform more complex iterations.
for i in range (3):
    for j in range (2):
        print(f"i:{i} and j:{j}")
```

```python
## Examples- calculate the sum of the first natural numbers using a while loop.
n = int(input("Enter the number:"))
sum = 0
count = 1
while count <= n:
    sum = sum + count
    count += 1
print(f"The sum of the first {n} natural numbers is: {sum}")
```

```python
## Using for loop 
n = int(input("Enter the number:"))
sum = 0
for i in range (1,n+1):
    sum = sum + i
print(f"The sum of the first {n} natural numbers is: {sum}")
```

```python
## Example- print all the prime numbers between 1 and 100.
 
for num in range (1,12):
    if num>1:
        for i in range (2,num):
            if num % i == 0:
                break
        else:
            print(num)    

```
