Variables


- Variables are the fundamental elements in programming used to store data can be referenced and manipulated in a program. In python, variables are created when you assign a value to them, and they do not require explicit declaration to reserve memory space.The declaration happens automatically when you assign a value to a variable.

```python
a = 100
print(a) # Output: 100
```

```python
## Decalring And Assigning Variables in Python
age = 20
height = 5.6
name = "Amiya"
is_student = True

##printing the variables
print("Age:", age) # Output: Age: 20
print("Height:", height) # Output: Height: 5.6
print("Name:", name) # Output: Name: Amiya
```

```python
##Naming Conventions for Variables in Python:
##1. Variable names must start with a letter (a-z, A-Z) or an underscore (_).
##2. Variable names can contain letters, digits (0-9), and underscores (_).
##3. Variable names are case-sensitive (age and Age are different variables).
##4. Variable names should not be a reserved keyword in Python (e.g., if, else, for, while, etc.).
##Examples of valid variable names:
first_name = "Amiya"
last_name = "Swain"
```

```python
##invalid variable names
#1. 2name = "Amiya" # Invalid: starts with a digit
#2. first-name = "Amiya" # Invalid: contains a hyphen
#3. class = "Python" # Invalid: 'class' is a reserved keyword
#4. name@ = "Amiya" # Invalid: contains a special character
```

```python
##case sensitivity of variable names
age = 20
Age = 30
age == Age # Output: False
```

```python
##Understanding Variable Types in Python
##Python is a dynamically typed language, which means that you don't need to declare the type of a variable when you create it. The type of a variable is determined at runtime based on the value assigned to it. Here are some common variable types in Python:
age = 25 # Integer
height = 5.8 # Float
name = "Amiya" # String
is_student = True # Boolean
print(type(age)) # Output: <class 'int'>
print(type(height)) # Output: <class 'float'>
print(type(name)) # Output: <class 'str'>
print(type(is_student)) # Output: <class 'bool'>
```

```python
## Type checking and conversion
type(age) # Output: <class 'int'>
type(height) # Output: <class 'float'>
age_str = str(age) # Convert integer to string
print(age_str) # Output: "25"
print(type(age_str)) # Output: <class 'str'>
```

```python
name = "Amiya"
int(name) # This will raise a ValueError since "Amiya" cannot be converted to an integer
```

```python
## Dynamic Typing in Python
##In Python, you can reassign a variable to a different type of value. This is because Python is dynamically typed, which allows for flexibility in variable usage. For example:
var = 10
print(var,type(var)) # Output: <class 'int'>
var = "Hello"
print(var,type(var)) # Output: <class 'str'>
var = 3.14
print(var,type(var)) # Output: <class 'float'>
```
